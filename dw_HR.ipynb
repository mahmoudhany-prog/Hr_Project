dim_Department

```python
import pandas as pd
from sqlalchemy import create_engine
import re


server = "localhost"
database = "cleaning_HR"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
cleaning_engine = create_engine(connection_string)

server = "localhost"
database = "dw_HR"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
dw_engine = create_engine(connection_string)


# Read from cleaning table

df = pd.read_sql("SELECT [Department] FROM [cleaning_HR].[dbo].[clean_HR]", cleaning_engine)
df.insert(0, "DepartmentID", range(1, len(df) + 1))

mapping = {
    'SALES': 'Business',
    'RESEARCH & DEVELOPMENT': 'Technical',
    'HUMAN RESOURCES': 'HR'
}

df['Department_Group'] = df['Department'].str.upper().map(mapping).fillna('Other')
df.to_sql("dim_Department", dw_engine, if_exists="replace", index=False)
print("Data cleaned and loaded into dw_HR!")
```

```python
import pandas as pd
from sqlalchemy import create_engine
import re


server = "localhost"
database ="cleaning_HR"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
cleaning_engine = create_engine(connection_string)

server = "localhost"
database = "dw_HR"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
dw_engine = create_engine(connection_string)

df=pd.read_sql("""SELECT 
       [Attrition]
      ,[DailyRate]
      ,[DistanceFromHome]
      ,[EmployeeCount]
      ,[EmployeeNumber] as[EmployeeID]
      ,[HourlyRate]
      ,[MonthlyIncome]
      ,[MonthlyRate]
      ,[NumCompaniesWorked]
      ,[Over18]
      ,[OverTime]
      ,[PercentSalaryHike]
      ,[StandardHours]
      ,[StockOptionLevel]
      ,[TotalWorkingYears]
      ,[TrainingTimesLastYear]
      ,[YearsAtCompany]
      ,[YearsInCurrentRole]
      ,[YearsSinceLastPromotion]
      ,[YearsWithCurrManager]
  FROM [cleaning_HR].[dbo].[clean_HR]""", cleaning_engine)
df.insert(0, "RoleID", range(1, len(df) + 1))
df.insert(0, "DepartmentID", range(1, len(df) + 1))
df.insert(0, "EducationID", range(1, len(df) + 1))
df.insert(0, "Satisfaction_ID", range(1, len(df) + 1))

df.to_sql("fact_HR", dw_engine, if_exists="replace", index=False)
print("Data inserted successfully into fact_HR table in the data warehouse.")
```
