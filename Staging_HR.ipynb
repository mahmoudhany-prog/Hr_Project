```python
import pandas as pd
from sqlalchemy import create_engine

# Connect to staging_db
server = "localhost"
database = "staging_HR"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
engine = create_engine(connection_string)

# Load Excel file
df = pd.read_excel(r"D:\ME\Hr Project\GBS BI HUB - BI Developer - HR Attrition Case Study.xlsx")

# Insert into staging table
df.to_sql("stg_HR", engine, if_exists="replace", index=False)
print("Excel data loaded into staging_HR!")
```
