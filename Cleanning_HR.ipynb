```python
import pandas as pd
from sqlalchemy import create_engine
import re

server = "localhost"
database = "staging_HR"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
staging_engine = create_engine(connection_string)

server = "localhost"
database = "cleaning_HR"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
cleaning_engine = create_engine(connection_string)

# Read from staging table
df = pd.read_sql("SELECT * FROM stg_HR", staging_engine)
# Clean data

df.drop_duplicates(subset=["EmployeeNumber"], inplace=True)
df["EmployeeNumber"]=df["EmployeeNumber"].astype(int)
df["Age"]=df["Age"].astype(int)
df["DailyRate"]=df["DailyRate"].astype(int)
df["HourlyRate"]=df["HourlyRate"].astype(int)
df["MonthlyIncome"]=df["MonthlyIncome"].astype(int)
df["PercentSalaryHike"]=df["PercentSalaryHike"].astype(int)


df.to_sql("clean_HR", cleaning_engine, if_exists="replace", index=False)
print("Data cleaned and loaded into cleaning_HR!")
```
